```python
# Import Libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn import preprocessing
```

```python
#Importing dataset
df=pd.read_csv('Company_Data.csv')
df
```

## EDA

```python
df.loc[df.Sales>=8.5,'sales']='High'
df.loc[df.Sales<8.5,'sales']='Low'
```

```python
df
```

```python
df.drop('Sales',inplace=True,axis=1)
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#finding Unique values and convert these catagorical values into numerical values
df.ShelveLoc.unique()
```

```python
df['ShelveLoc']=preprocessing.LabelEncoder().fit_transform(df['ShelveLoc'])
```

```python
df.Urban.unique()
```

```python
df['Urban']=preprocessing.LabelEncoder().fit_transform(df['Urban'])
df.US.unique()
```

```python
df['US']=preprocessing.LabelEncoder().fit_transform(df['US'])
```

```python
df.sales.unique()
```

```python
df['sales']=preprocessing.LabelEncoder().fit_transform(df['sales'])
```

```python
df
```

```python
#devide the data into target and predectors
columns=list(df.columns)
predictors=columns[0:10]
target=columns[-1]
```

```python
predictors
```

```python
target
```

```python
X=df[predictors]
Y=df[target]
```

## Model Building

```python
RF=RandomForestClassifier(n_jobs=2,oob_score=True,n_estimators=1000,criterion='entropy')
RF
```

## Model Prediction

```python
RF.fit(X,Y)
```

```python
RF.estimators_
```

```python
RF.classes_
```

```python
RF.n_features_
```

```python
RF.n_outputs_
```

```python
RF.n_outputs_
```

```python
df['rf_redict']=RF.predict(X)
```

```python
confusion_matrix(df['sales'],df['rf_redict'])
```

## Now will try 500 decision trees

```python
RF=RandomForestClassifier(n_jobs=2,n_estimators=500,oob_score=True,criterion='entropy')
```

```python
RF.fit(X,Y)
```

```python
df['rf_predict2']=RF.predict(X)
```
